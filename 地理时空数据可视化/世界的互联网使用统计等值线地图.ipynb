```python
import pandas as pd
import plotly
import plotly.express as px

internet_usage_df = pd.read_csv('data/share-of-individuals-using-the-internet.csv')
internet_usage_df.head(5)
```

```python
fig = px.choropleth(
    internet_usage_df,
    locations="Code",
    color="Individuals using the Internet (% of population)",
    hover_name="Country")

fig.update_layout(
    title_text='Internet usage across the world (% population)')

fig.show()
plotly.offline.plot(
    fig,
    filename='result/Internet usage across the world (% population) flat.html')
```
