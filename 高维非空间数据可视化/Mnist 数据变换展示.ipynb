```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import MDS, Isomap, LocallyLinearEmbedding, TSNE
```

```python
""" MDS"""

n_points = 2000
X, color = datasets.make_swiss_roll(n_points, random_state=0)

fig = plt.figure(figsize=(45, 25))
ax = fig.add_subplot(251, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap='jet', s=9, lw=1)
ax.view_init(10, -72)  # type: ignore

#MDS
mds = MDS(n_components=2)
X_mds = mds.fit_transform(X)

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(X_mds[:, 0], X_mds[:, 1], c=color, cmap='jet', s=9, lw=1)
```

```python
""" ISOMAP"""

digits = datasets.load_digits(n_class=6)
X = digits.data
y = digits.target
n_samples, n_features = X.shape
print(n_features)
print(n_samples)

isomap = Isomap(n_components=2, n_neighbors=10)
X_isomap = isomap.fit_transform(X)

plt.figure(figsize=(10, 10))
plt.scatter(X_isomap[y == 0, 0], X_isomap[y == 0, 1], color='blue', alpha=0.5, label='0', s=9, lw=2)
plt.scatter(X_isomap[y == 1, 0], X_isomap[y == 1, 1], color='green', alpha=0.5, label='1', s=9, lw=2)
plt.scatter(X_isomap[y == 2, 0], X_isomap[y == 2, 1], color='orange', alpha=0.5, label='2', s=9, lw=2)
plt.scatter(X_isomap[y == 3, 0], X_isomap[y == 3, 1], color='purple', alpha=0.5, label='3', s=9, lw=2)
plt.scatter(X_isomap[y == 4, 0], X_isomap[y == 4, 1], color='violet', alpha=0.5, label='4', s=9, lw=2)
plt.scatter(X_isomap[y == 5, 0], X_isomap[y == 5, 1], color='red', alpha=0.5, label='5', s=9, lw=2)
```

```python
""" LLE"""

digits = datasets.load_digits(n_class=6)
X = digits.data  # type: ignore
y = digits.target  # type: ignore
n_samples, n_features = X.shape
print(n_features)
print(n_samples)

lle = LocallyLinearEmbedding(n_neighbors=10, n_components=2,
                             method='standard')
X_lle = lle.fit_transform(X)

plt.figure(figsize=(10, 10))
plt.scatter(X_lle[y == 0, 0], X_lle[y == 0, 1], color='blue', alpha=0.5, label='0', s=9, lw=2)
plt.scatter(X_lle[y == 1, 0], X_lle[y == 1, 1], color='green', alpha=0.5, label='1', s=9, lw=2)
plt.scatter(X_lle[y == 2, 0], X_lle[y == 2, 1], color='orange', alpha=0.5, label='2', s=9, lw=2)
plt.scatter(X_lle[y == 3, 0], X_lle[y == 3, 1], color='purple', alpha=0.5, label='3', s=9, lw=2)
plt.scatter(X_lle[y == 4, 0], X_lle[y == 4, 1], color='violet', alpha=0.5, label='4', s=9, lw=2)
plt.scatter(X_lle[y == 5, 0], X_lle[y == 5, 1], color='red', alpha=0.5, label='5', s=9, lw=2)
```

```python
""" t-SNE """

digits = datasets.load_digits(n_class=6)
X = digits.data  # type: ignore
y = digits.target  # type: ignore
n_samples, n_features = X.shape
print(n_features)
print(n_samples)

tsne = TSNE(n_components=2, random_state=0)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(10, 10))
plt.scatter(X_tsne[y == 0, 0], X_tsne[y == 0, 1], color='blue', alpha=0.5, label='0', s=9, lw=1)
plt.scatter(X_tsne[y == 1, 0], X_tsne[y == 1, 1], color='green', alpha=0.5, label='1', s=9, lw=1)
plt.scatter(X_tsne[y == 2, 0], X_tsne[y == 2, 1], color='orange', alpha=0.5, label='2', s=9, lw=1)
plt.scatter(X_tsne[y == 3, 0], X_tsne[y == 3, 1], color='purple', alpha=0.5, label='3', s=9, lw=1)
plt.scatter(X_tsne[y == 4, 0], X_tsne[y == 4, 1], color='violet', alpha=0.5, label='4', s=9, lw=1)
plt.scatter(X_tsne[y == 5, 0], X_tsne[y == 5, 1], color='red', alpha=0.5, label='5', s=9, lw=1)
```
